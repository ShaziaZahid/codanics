```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# seaborn comes with 17 datasets according to documnetation but these are more than 17.
#Each item in this list maps to a dataset on the repo : for example, 
#the dataset named `iris` maps to the `iris.csv` file.

sns. get_dataset_names() 
```

```python
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
df.describe()
```

```python
# groupby is same as SQL groupby, find all unique in column and then make groups relating to each.
df.groupby('survived').describe()
```

```python
#multiple columns group by (will follow order)
df.groupby(['survived', 'sex']).describe()
```

```python
df.groupby(['survived', 'sex']).count()
```

```python
df.groupby(['survived', 'sex', 'who']).count()
```
