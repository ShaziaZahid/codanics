```python
# to create a series of dates starting from given date and going to 6
mydates = pd.date_range("20130101", periods=6)

mydates
```

```python
# creating a data frame, using random function of numpy to create values, 
#indexes are dates(mydates creaded above) a sereis
# column names are from list (ABCD)
df = pd.DataFrame(np.random.randn(6, 4), index=mydates, columns=list("ABCD"))

df
```

```python
# creating a dataframe from dictionary
# keys are ABCD but values are varing
# getting four rows because of of categoreis in E (test,train,test, train)
df2 = pd.DataFrame(
    {
        "A": 1.0,
        "B": pd.Timestamp("20130102"),
        "C": pd.Series(1, index=list(range(4)), dtype="float32"),
        "D": np.array([3] * 4, dtype="int32"),
        "E": pd.Categorical(["test", "train", "test", "train"]),
        "F": "foo",
    }
)


df2
```
